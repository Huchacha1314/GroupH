```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = np.linspace(-1, 1, num =100)
alpha_values = np.where(alpha < 0, np.nan, alpha)
x = np.sqrt((alpha_values - 0)/1)
start_alpha = 0
pos_equli = x
neg_equli = -x
```

```python
plt.plot(alpha, pos_equli)
plt.plot(alpha, neg_equli)
plt.plot(start_alpha, 0, '.')
plt.legend()
plt.show()
```
