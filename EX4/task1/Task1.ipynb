```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from utils import *

%load_ext autoreload
%autoreload 2
```

```python
# define notebook parameters
w = 3
Y, X = np.mgrid[-w:w:100j, -w:w:100j]

y0 = np.array([3, 3])
time = np.linspace(0, 10, 100)

# example matrix
```

# unstable focus

```python
alpha = 0.1
A = np.array([
    [alpha, alpha],
    [-1/4, 0]
])
eigenvalues = np.linalg.eigvals(A)
yt, time = solve_euler(lambda y: A@y, y0, time)

# example linear vector field A*x
ax0 = plot_phase_portrait(A, X, Y)
plt.title("\u03B1: " + str(alpha) + ", \u03BB_1: " + str(eigenvalues[0]) + ", \u03BB_2: " + str(eigenvalues[1])+"--unstable focus")


plt.savefig("unstable_focus")
```

# unstable node

```python
alpha = 2
A = np.array([
    [alpha, alpha],
    [-1/4, 0]
])
eigenvalues = np.linalg.eigvals(A)
yt, time = solve_euler(lambda y: A@y, y0, time)

# example linear vector field A*x
ax0 = plot_phase_portrait(A, X, Y)
plt.title("\u03B1: " + str(alpha) + ", \u03BB_1: " + str(np.around(eigenvalues[0],2)) + ", \u03BB_2: " + str(np.around(eigenvalues[1],2)) + " -- unstable node")

ax0.set_aspect(1)
plt.savefig("unstable_node")
```

# stable focus

```python
alpha = 0.1

A = np.array([
    [-alpha, alpha],
    [-1/4, 0]
])
eigenvalues = np.linalg.eigvals(A)
yt, time = solve_euler(lambda y: A@y, y0, time)

# example linear vector field A*x
ax0 = plot_phase_portrait(A, X, Y)
plt.title("\u03B1: " + str(alpha) + ", \u03BB_1: " + str(np.around(eigenvalues[0],2)) + ", \u03BB_2: " + str(np.around(eigenvalues[1],2))+"--stable focus")

ax0.set_aspect(1)
plt.savefig("stable_focus")
```

# stable node 

```python
alpha = 2
A = np.array([
    [-alpha, alpha],
    [-1/4, 0]
])
eigenvalues = np.linalg.eigvals(A)
yt, time = solve_euler(lambda y: A@y, y0, time)

# example linear vector field A*x
ax0 = plot_phase_portrait(A, X, Y)
plt.title("\u03B1: " + str(alpha) + ", \u03BB_1: " + str(np.around(eigenvalues[0],2)) + ", \u03BB_2: " + str(np.around(eigenvalues[1],2))+"--stable node")

ax0.set_aspect(1)
plt.savefig("stable_node")
```
